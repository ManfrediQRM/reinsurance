```python
#QUOTA SHARE REINSURANCE
class QuotaShare(object):
    def __init__(self, data_file_name):
        import pandas as pd
        import numpy as np 

        np.random.seed(1)
        
        self.data = pd.read_csv(data_file_name)
        self.data = self.data[["ID","CLM_AMT"]]
        self.data = self.data.replace({"\$":"",",":""},regex=True).astype(int) #trasformo in integer e rimuovo simbolo dollaro
        self.data["PREM"] = np.random.randint(low=500, high=4000,size = len(self.data))
    
    def apply_treaty_structure(self,cession_rate,commission_rate,rows):
        self.data["CLM_CEDED"] = self.data["CLM_AMT"]* cession_rate#nuova colonna creata. amount paid by reinsurer, esempio 40% quota share allora il reinsurer pagherà il 40% del claim
        self.data["PREM_CEDED"] = self.data["PREM"]* cession_rate # nuova colonna creata
        self.data["CEDING_COMMISSION"] = self.data["PREM_CEDED"] * commission_rate # nota che il commission rate si applica al PREMIUM CEDED
        self.data["CLM_NET"] = self.data["CLM_AMT"]* (1-cession_rate) #what the primary insurance company retains, ovvero net claim amount
        self.data["PREM_NET"] = self.data["PREM"] * (1-cession_rate) + self.data["CEDING_COMMISSION"]

        return self.data.head(rows)


        
```

```python
qs = QuotaShare("/Users/manfredi/Desktop/python-introduction-main/actuarial/dati/car_insurance_claim.csv")
print(qs.data.head(10)) # every policy is gonna pay a premium, but not every policy will have a claim
qs.apply_treaty_structure(0.4,0.22,10) # immagina cession rate = 22% e vogliamo 10 righe
```
